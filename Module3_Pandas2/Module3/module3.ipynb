1. Водителей какого пола в датасете больше?

```python
import pandas as pd

data = pd.read_csv("police.csv", sep=",")


countsNull = data.isnull().sum()


max_null_column = countsNull.idxmax()
# max_null_count = countsNull.max()

data.drop(max_null_column, inplace=True, axis=1)
print(data.driver_gender.str.contains("M").value_counts())
gender_counts = data.driver_gender.value_counts()
print(f"Количество мужчин: {gender_counts.get('M', 0)}")
print(f"Количество женщин: {gender_counts.get('F', 0)}")
```

Чтобы получить ответ на вопрос, водителей какого пола чаще останавливают

```python
data['driver_gender'].value_counts(normalize=True)
```
