```python
import spacy
import numpy as np
import en_core_web_sm


def cosine(v1, v2):
    if np.linalg.norm(v1)*np.linalg.norm(v2) > 0:
        return np.dot(v1, v2)/(np.linalg.norm(v1)*np.linalg.norm(v2))
    else:
        return 0

nlp = spacy.load('en_core_web_sm')
word_1 = 'cat'
word_2 = 'cow'

cosine(nlp(word_1).vector, nlp(word_2).vector)
```

Анализ похожих товаров по их описанию
Допустим, вы аналитик данных в компании, которая занимается продажей мебели. Ваша задача — определить, какие товары наиболее похожи друг на друга по описанию. Для этого необходимо использовать косинусную меру угла с помощью библиотеки spacy.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import spacy


data = pd.read_csv('product_description.csv')

nlp = spacy.load("en_core_web_sm")

def preprocess_text(text):
    doc = nlp(text)
    tokens = [token.lemma_.lower() for token in doc if not token.is_stop and not token.is_punct]
    return " ".join(tokens)
    
data['processed_text'] = data['description'].apply(preprocess_text)


print(data.info())
print(data.head(10))

vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform(data['processed_text'])

# Рассчитываем косинусную схожесть между всеми парами товаров
cos_sim_matrix = cosine_similarity(vectors)

# Преобразуем матрицу схожести в DataFrame для удобства
cos_sim_df = pd.DataFrame(cos_sim_matrix, columns=data['product_name'], index=data['product_name'])

# Для каждого товара, находим 5 наиболее похожих товаров (без учета самого себя)
top_5_similar = {}
for product in cos_sim_df.columns:
    top_5_similar[product] = cos_sim_df[product].sort_values(ascending=False).iloc[1:6]
    print(product)


print(top_5_similar)
```
